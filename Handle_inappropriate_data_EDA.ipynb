First Method of Removing the Negative Values(Using .loc )

```python
df3.loc[df3["Rating(1-5)"]<0 ,"Rating(1-5)"]="NAN"
```

```python
df3
```

Second method for Removing the negative values (Using Numpy)


```python
# Another way of removing negative value
df3["NoOfPax"]=np.where(df3["NoOfPax"]<0,"NAN", df3["NoOfPax"])
```

```python
df3
```

Third method of removing negative value (Using Masking Function)

```python
df3["EstimatedSalary"].mask(df3["EstimatedSalary"]<0,"NAN",inplace=True)
df3
```

```python
df3["Hotel"]=np.where(df3["Hotel"]=="Ibys","Ibis",df3["Hotel"])
```
